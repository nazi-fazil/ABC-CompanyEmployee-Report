### 5.Find if there is any correlation between age and salary , represent it visually.¶

```python
# Group by age category and find the average salary for each category
average_salary_by_category = employee_df.groupby('AgeCategory',observed=False)['Salary'].mean()
```

```python
highest_salary_category = average_salary_by_category.idxmax()
highest_salary_value = average_salary_by_category.max()

print(f"The age category with the highest average salary is {highest_salary_category} with an average salary of {highest_salary_value:.2f}")
```

```python
sns.barplot(x=average_salary_by_category.index,y=average_salary_by_category.values,hue=average_salary_by_category.index,legend=False,palette='viridis')
plt.xlabel('Age Category')
plt.ylabel('Average Salary')
plt.title('Average Salary by Age Category')
plt.show()
```

#### **Data Story**
Upon analyzing the barplot, it becomes evident that the age group 31-35 commands the highest average salary among employees. Conversely, the 18-20 age category records the lowest average salary. It is noteworthy that the salary distribution is not strictly proportional to age, as the presence of the 36-40 age category implies. This suggests that factors beyond age alone contribute to the variations in salary levels among employees.
